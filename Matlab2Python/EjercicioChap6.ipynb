#Datos originales

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.interpolate
```

```python
x=np.array([0,0.5,1.0,1.5])
y=np.array([1.0,2.0,1.5,1.0])
plt.plot(x,y,'*r')
plt.title('Datos originales')
plt.axis([0,1.8,0,2.4])
plt.show()
```

#Interpolacion vandermode#
Sacado de : https://jyyuan.wordpress.com/2014/01/02/polynomial-interpolation-using-vandermonde-matrix-and-least-squares/

```python

n=51
start=0
finish=2
order=4 
mat_dim=(len(x),order) #una parabola
A=np.fliplr(np.vander(x,mat_dim[1]))
coefs,_,_,_=np.linalg.lstsq(A,y)
```

```python
interp_x=np.linspace(start,finish,n)
interp_y=np.zeros(n)
for ind,ix in enumerate(interp_x):
    interp_y[ind]=np.sum(coefs*ix**np.arange(0,order))
```

```python
plt.plot(interp_x,interp_y,'-b',label='interp line')
plt.plot(x,y,'*r',label='data points')
plt.title('Vandermode')
plt.axis([0,1.8,0,2.4])
plt.show()
```

#Lagrange interpolation

```python

poly=scipy.interpolate.lagrange(x,y)
y_lang=poly(interp_x)
```

```python
plt.plot(interp_x,y_lang,'-b',label='interp line')
plt.plot(x,y,'*r',label='data points')
plt.title('Lagrange')
plt.axis([0,1.8,0,2.4])
plt.show()
```

#6.2 interpolacion de tablas dinamicas

```python
temp=np.array([100.0,110.0,130.0,150.0])#En celsius
presion=np.array([101.33,143.27,270.1,475.8])#Presion kpa
poly2=scipy.interpolate.lagrange(temp,presion)
x2=np.linspace(100,150,51)
y2=poly2(x2)

plt.plot(x2,y2,'-b',label='interp line')
plt.plot(temp,presion,'*r',label='data points')
plt.title('Grafica Termodinamica')
plt.xlabel('Temperatura [C]')
plt.ylabel('Presion [kPa]')
plt.show()
```

Los valores de presion para los siguientes valores de temperatura

```python
print(poly2(112))
print(poly2(114))
print(poly2(120))
print(poly2(136))
print(poly2(148))
```
