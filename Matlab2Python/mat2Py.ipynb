```python
import numpy as np 
from matplotlib import pyplot as plt
```

2.4 Arreglos y opreaciones sobre arreglos

```python
x=np.pi*np.linspace(0,1,11)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Sin(x)')
plt.title("Lo maximo")
plt.show()
```
